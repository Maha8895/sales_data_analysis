```python
df=pd.read_csv('Walmart.csv', encoding_errors='ignore')
```

```python
# To get the statistics information of the dataset
df.describe
```

```python
# To get the datatypes information of the columns
df.info()
```

```python
# To find all the duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.shape
```

```python
# To find the NULL values

df.isnull().sum()
```

```python
#Dropping all rows with NULL values,because we're not sure about the unit price and quantity.

df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Since unit_price is a string, we need to replace $ with nothing and then typecast to float

df['unit_price']=df['unit_price'].str.replace('$','')
```

```python
df.head()
```

```python
df['unit_price']=df['unit_price'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
#Extract Weekday from Date
df['date'] = pd.to_datetime(df['date'])
df['weekday']=df['date'].dt.day_name()
df.head()
```

```python
#Extract Hour from Time
df['time']=pd.to_datetime(df['time'],format='%H:%M:%S')
df['hour']=df['time'].dt.hour
df.head()
```

```python
df.columns=df.columns.str.lower()
```

```python
df['total_amount']= df['unit_price']* df['quantity']
df.head()
```

```python
# rounding up the total value to 1 decimal place
import numpy as np
df['total_amount']=np.ceil(df['total_amount']*10)/10
df.head()
```
